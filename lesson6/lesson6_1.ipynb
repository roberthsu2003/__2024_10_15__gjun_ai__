```python
#小明想要存錢買一輛機車,機車每輛30000元，他將每月存的錢輸入，當存款足夠買機車時，就顯示提示訊息告知。
deposit = 0
month = 0

while(True):
    month += 1
    money = int(input(f"請輸入第{month}月份的存款:"))
    deposit += money
    if deposit >= 30000:
        break
print(f"恭喜!已經存夠了，存了{month}個月的總存款為:{deposit}元。")
```

```python
#請輸入小寫英文字,輸出第一個字大寫的英文字,如果按q英文字,將離開應用程式

while(True):
    stuff = input("請輸入小寫英文字[按q會離開]:")
    if stuff.upper() == 'Q':
        break
    print(stuff.capitalize())
print("應用程式結束")
```

```python
'''
docstring
顯示:
請輸入整數,輸入[q]離開
輸入的格式錯誤,要求重新輸入
奇數輸出平方
偶數不做任何動作
'q'要離開
'''

while(True):
    value = input("請輸入整數,輸入[q]離開:")
    if value.upper() == 'Q':
        break
    try:
        value_int = int(value)
    except:
        print("請重新輸入")
        continue
    if not (value_int % 2) == 0:
        print(value_int**2)
    
print("應用程式結束")
```
