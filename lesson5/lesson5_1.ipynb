```python
chinese = int(input("請輸入國文分數"))
math = int(input("請輸入數學分數"))

if chinese == 100:
    if math == 100:
        bonus = 1000
    else:
        bonus = 500
elif math == 100:
    bonus = 500
else:
    bonus = 0

print(f"獎金:{bonus}")
    
```
