#### if...else...

```python
scores = int(input("請輸入學生分數0~100"))
if scores >= 60:
    print("及格")
    if scores >= 90:
        print("優")
    else:
        print("一般")
else:
    print("不及格")
    if scores >= 50:
        print("丙")
    else:
        print("丁")
```

```python
scores = int(input("請輸入學生分數0~100"))
if scores >= 90:
    print('優')
else:
    if scores >= 80:
        print('甲')
    else:
        if scores >= 70:
            print('乙')
        else:
            if scores >= 60:
                print('丙')
            else:
                print('丁')
```

```python
#多項選擇
scores = int(input("請輸入學生分數0~100"))
if scores >= 90:
    print('優')
elif scores >= 80:    
    print('甲')
elif scores >= 70:
    print('乙')
elif scores >= 60:            
    print('丙')
else:
    print('丁')
```
