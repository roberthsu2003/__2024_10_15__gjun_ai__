#### 判斷if

```python
#學生總分最高為300
#如果分數超過300,就以300分為準
#有些學生可以加分5%
scores = eval(input('請輸入學生分數(最高300分):'))
if(scores > 300):
    scores = 300
is_add = input("學生是否符合加分條件?(y,n)")
if is_add == 'y':
    scores *= 1.05

print(f"學生的分數:{scores:.0f}")
```

```python
#學生總分最高為300
#如果分數超過300,就以300分為準
#有些學生可以加分5%
scores = eval(input('請輸入學生分數(最高300分):'))
if(scores > 300):
    scores = 300
is_add = input("學生是否符合加分條件?(y,n)")
if is_add == 'y': #巢狀判斷
    scores *= 1.05
    if scores > 300:
        scores = 300
print(f"學生的分數:{scores:.0f}")
```
