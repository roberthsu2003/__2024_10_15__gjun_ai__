```python
import matplotlib.pyplot as plt

# 1.製作figure
fig = plt.figure()

# 2.在1.的figure上製作一個axes
axes = fig.add_subplot(1, 1, 1)

# 3.將圖的數據儲存在axes中
X = [0, 1, 2]
Y = [3, 4, 5]
axes.plot(X, Y,'ro-')

# 4.顯示圖
plt.show()
```

```python
import matplotlib.pyplot as plt

figure = plt.figure()
axes_1 = figure.add_subplot(1,3,1)
axes_2 = figure.add_subplot(1,3,2)
axes_3 = figure.add_subplot(1,3,3)
axes_1.set_title('1')
axes_2.set_title('2')
axes_3.set_title('3')
plt.show()
```
