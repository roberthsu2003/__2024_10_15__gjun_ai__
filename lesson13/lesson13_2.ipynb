```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-3, 3, num=50)
y1 = x * 2 + 1
y2 = x ** 2

figure = plt.figure()
axes = figure.add_subplot()
axes.plot(x,y1,color='red',linestyle='--')
axes.plot(x,y2,color='green',linestyle='-.')
axes.set_xlim(-1, 2)
axes.set_ylim(-2, 3)
plt.show()
```
